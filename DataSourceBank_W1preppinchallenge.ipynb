```python
import pandas as pd
```

```python
df = pd.read_csv('data/transactions.csv')
```

```python
df.head(5)
```

```python
df['Bank'] = df['Transaction Code'].str.split("-").str[0]
df['Bank']
```

```python
df.head(2)
```

```python
df['Online or In-Person']
```

```python
def online_inline(values):
    if values == 1:
        return 'Online'
    elif values == 2:
        return 'In-person'
```

```python
df['Online or In-Person'] = df['Online or In-Person'].apply(online_inline)
```

```python
df.head(2)
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'],dayfirst=True)
```

```python
df['Transaction Date'] = df['Transaction Date'].dt.day_name()
```

```python
df.head(2)
```

```python
# Total Values of Transactions by each bank
totalvalue_per_bank = df.groupby('Bank')['Value'].sum().reset_index()
```

```python
totalvalue_per_bank = totalvalue_per_bank.sort_values(by='Value', ascending=False)
totalvalue_per_bank
```

```python
# Total Values by Bank, Day of the Week and Type of Transaction
```

```python
totalvalue_per_bank_day_type = df.groupby(['Bank','Transaction Date','Online or In-Person'])['Value'].sum().reset_index()
```

```python
totalvalue_per_bank_day_type 
```

```python
# Total Values by Bank and Customer Code
totalvalue_perbank_code = df.groupby(['Bank','Customer Code'])['Value'].sum().reset_index()
totalvalue_perbank_code
```
